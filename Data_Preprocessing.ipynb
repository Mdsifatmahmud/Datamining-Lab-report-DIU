```python
Q1 = df['avg_glucose_level'].quantile(0.25)
Q3 = df['avg_glucose_level'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier = df[(df['avg_glucose_level'] < lower_bound) | (df['avg_glucose_level'] > upper_bound)]
print(len(outlier))
```

```python
df = df[(df['avg_glucose_level'] >= lower_bound) & (df['avg_glucose_level'] <= upper_bound)]
print(len(df))
```

```python
x = df.drop('stroke',axis=1)
y = df['stroke']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_features = SelectKBest(score_func = f_classif)
fit_features.fit(x,y)
```

```python
fs = pd.DataFrame(fit_features.scores_,index=x.columns, columns = ['score values'])
fs
```

```python
fs.nlargest(10,'score values')
```

```python
fs.nlargest(15, 'score values').plot(kind="barh", figsize=(10, 5), color='orange', edgecolor='black')
plt.title('Top 15 Features by Score Value', fontsize=16)
plt.xlabel('Score Value', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
x = df[fs[fs['score values'] > 2.5].index]
x
```
